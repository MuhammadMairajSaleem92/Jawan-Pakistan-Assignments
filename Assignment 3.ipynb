### Question no. 1 : Create a new column in the provided dataset. Column name should be num_of_cast. The assignment was already explained in class. You have to append or add a column in the dataset which contains the total number of cast in that particular movie or Tv-Show.

```python
data['cast'] = data['cast'].replace(np.nan, 'None')
data.tail()
```

```python
cast_length_list = []
for i in range(0,7787):
    if data.loc[i, 'cast'] == 'None':
        cast_length = 0
        cast_length_list.append(cast_length)
    else:
        cast_length = len(data.loc[i, 'cast'].split(', '))
        cast_length_list.append(cast_length)
```

```python
cast_length_list
```

```python
data['num_of_cast'] = cast_length_list
data.tail()
```
