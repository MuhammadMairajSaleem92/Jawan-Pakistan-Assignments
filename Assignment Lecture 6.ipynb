### Q1: Create High_Low_Speed Column

```python
# First find the mean
Speed_mean = data.Speed.mean()

# Function for Speed
def high_low_speed(val):
    if val < Speed_mean:
        return "Low Speed"
    elif val == Speed_mean:
        return "Neutral Speed"
    else:
        return "High Speed"
```

```python
# add the column of High_Low_Speed
data['High_Low_Speed'] = data.Speed.apply(high_low_speed)
data.head()
```

```python
# Summary of Count in High_Low_Speed for each category
data.High_Low_Speed.value_counts()
```
